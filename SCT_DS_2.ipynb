**Dataset:** Titanic Dataset (Kaggle)

**Objective:** To clean and analyze the Titanic dataset using exploratory data analysis techniques to identify important factors affecting passenger survival and uncover patterns, trends, and relationships among variables.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
# Shape of dataset
df.shape
```

```python
# Column names and data types
df.info()
```

```python
# Statistical summary
df.describe()
```

```python
# Check missing values
df.isnull().sum()
```

**Data Cleaning**

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
#removing duplicates
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

**Exploratory Data Analysis(EDA)**

```python
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.xticks([0,1],['Died','Survival'])
plt.savefig("Survival_count.png")
plt.show()

#percent table
survival_percent = df['Survived'].value_counts(normalize=True) * 100
survival_percent
```

```python
plt.figure(figsize=(6,4))
ax = sns.countplot(x='Sex', hue='Survived', data=df)

total = len(df)

for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%'
    ax.annotate(
        percentage,
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center',
        va='bottom'
    )

plt.title('Survival by Sex (Count & Percentage)')
plt.xticks([0,1],['Died','Survival'])
plt.ylabel('Passenger Count')
plt.xlabel('Sex')
plt.savefig("Survival_by_Sex.png")
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Passenger Class')
plt.savefig("Survival_by_Passenger_Class.png")
plt.show()

#percentage
class_survival_percent = pd.crosstab(df['Pclass'], df['Survived'], normalize='index') * 100
class_survival_percent
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.savefig("Age_Distribution.png")
plt.show()
```

```python
sns.countplot(x='Sex', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.savefig("Gender_Distribution.png")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age Distribution by Survival')
plt.xlabel('Survival Status (0 = Not Survived, 1 = Survived)')
plt.ylabel('Age')
plt.savefig("Age_Distribution_by_Survival.png")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.histplot(df['Fare'], bins=40, kde=True)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.savefig("Fare_Distribution.png")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare Distribution by Survival ')
plt.xlabel('Survival Status (0 = Not Survived, 1 = Survived)')
plt.ylabel('Fare')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numeric_df.corr()
correlation_matrix

plt.figure(figsize=(10,6))
sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    linewidths=0.5
)
plt.title('Numeric Correlation Matrix')
plt.savefig("Numeric_Correlation_Matrix.png")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title('Survival by Embarked Port')
plt.xlabel('Embarked Port')
plt.ylabel('Passenger Count')
plt.savefig("Survival_by_Embarked_Port.png")
plt.show()
```

```python
# Family size = SibSp + Parch + 1 (including the passenger)
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df[['SibSp', 'Parch', 'FamilySize']].head()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x='FamilySize', hue='Survived', data=df)
plt.title('Survival by Family Size')
plt.xlabel('Family Size')
plt.ylabel('Passenger Count')
plt.savefig("Survival_by_Family_Size.png")
plt.show()
```

```python
pivot = pd.pivot_table(df, values='Survived', index='Pclass', columns='Sex', aggfunc='mean')
pivot = (pivot * 100).round(2)
display(pivot)
```

**Final Observations**

1.The overall survival rate on the Titanic was low, with less than half of the passengers surviving, indicating the severity of the disaster.

2.Gender was the most influential factor, as female passengers had a significantly higher survival rate than male passengers.

3.Passenger class strongly affected survival, with first-class passengers showing much higher survival rates than second- and third-class passengers.

4.Fare had a positive relationship with survival, suggesting that higher-paying passengers (often in higher classes) had better access to safety measures.

5.Age showed a moderate impact on survival, with children having better survival chances compared to adults and senior passengers.

6.Family size influenced survival outcomes, where passengers traveling in small families survived more than solo travelers or very large families.

7.Correlation analysis confirmed that survival depended on multiple interacting factors, including class, fare, gender, age, and family structure, rather than any single variable alone.
